## Can we accurately predict which type of crime is committed? (Crime classification task)

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

crime_df['DAY_OF_WEEK'] = label_encoder.fit_transform(crime_df['DAY_OF_WEEK'] ) 
crime_df['STREET'] = label_encoder.fit_transform(crime_df['STREET'])
# crime_df['OFFENSE_CODE_GROUP'] = label_encoder.fit_transform(crime_df['OFFENSE_CODE_GROUP'])
crime_df['DISTRICT'] = label_encoder.fit_transform(crime_df['DISTRICT'])
crime_df.head()
```

```python
district_labels = label_encoder.inverse_transform(crime_df["DISTRICT"])
district_labels = sorted(pd.unique(district_labels))
print(district_labels)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = crime_df.drop(['DISTRICT'] , axis=1)
Y = crime_df.DISTRICT

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33, random_state=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```
